```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
import numpy as np
x = np.linspace(0,5,11)
y = x**2
```

```python
x
```

```python
y
```

```python
#functional plotting method
plt.plot(x,y)
```

```python
plt.plot(x,y,'r')
```

```python
plt.plot(x,y)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('y=x*x')
```

```python
#object oriented method
fig = plt.figure()
axes = fig.add_axes([0.1,0.1,0.8,0.8])
axes.plot(x,y)
#all in one line works too
#plt.figure().add_axes([0.1,0.1,0.8,0.8]).plot(x,y)

axes.set_xlabel('X-axis')
axes.set_ylabel('Y-axis')
axes.set_title('Title')
```

```python
fig = plt.figure()
axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes2 = fig.add_axes([0.2,0.5,0.4,0.3])
axes1.plot(x,y)
axes2.plot(y,x)
```

```python
#figure size and dpi
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,x,label='x')
ax.plot(x,x**2,label='x*x',linestyle='steps')
ax.plot(x,x**3,label='x*x*x',marker='o')
ax.legend()
```

```python
#to save the plot
#fig.savefig('a.pdf',dpi=1000)
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.set_xlim([0,3])
ax.set_ylim([0,80])
ax.plot(x,x,label='x')
ax.plot(x,x**2,label='x*x',linestyle='steps')
ax.plot(x,x**3,label='x*x*x',marker='o')
ax.legend()
```

```python
#scatter plot
#better in seaborne
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.scatter(x,x**2)
```
